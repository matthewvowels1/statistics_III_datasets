```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python


def generate(N):
    Ux = np.random.randn(N)
    Uy = np.random.randn(N)
    
    y = 0.3 + 0.15*Ux - 0.3*Ux**2 + Uy
    
    return y, Ux
```

```python
N = 400

y, x = generate(N)

df = pd.DataFrame(data=(y,x, x**2)).T
df.columns = ['y', 'x', 'x2']
df.to_csv('quadratic.csv', index=False)

plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

print(df.head())

```
